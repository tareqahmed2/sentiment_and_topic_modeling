```python
# =========================================
# CREATE FOLDER
# =========================================
import os
import pandas as pd

output_folder = "graphData"

if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
# =========================================
# 2. LOAD DATA
# =========================================
comments = pd.read_csv("Data/English_version/English_translated_comment_data.csv")
live = pd.read_csv("Data/English_version/English_translated_live_chat_data.csv")
videos = pd.read_csv("Data/English_version/English_translated_video_informationt.csv")
```

```python
# =========================================
# ADD CHANNEL TO COMMENTS
# =========================================
comments = comments.merge(
    videos[['video_id','channel']],
    on='video_id',
    how='left'
)

```

```python
# =========================================
# CREATE USER NODES
# =========================================
users = pd.concat([
    comments[['user_id']],
    live[['user_id']]
]).drop_duplicates()

users['label'] = "User"

```

```python
# =========================================
# CREATE VIDEO NODES
# =========================================
video_nodes = videos[['video_id','channel','Title']]
video_nodes['label'] = "Video"
```

```python
# =========================================
# CREATE COMMENT NODES
# =========================================
comments['comment_id'] = "c_" + comments.index.astype(str)

comment_nodes = comments[['comment_id','translated_text']]
comment_nodes['type'] = "Comment"
```

```python
# =========================================
# CREATE LIVE CHAT NODES
# =========================================
live['live_id'] = "l_" + live.index.astype(str)

live_nodes = live[['live_id','translated_text','channel']]
live_nodes['type'] = "LiveChat"
```

```python
# =========================================
# RELATIONSHIPS
# =========================================

# User -> Comment
user_comment = comments[['user_id','comment_id']]
user_comment.columns = ['source','target']
user_comment['type'] = "POSTED"

# User -> Live
user_live = live[['user_id','live_id']]
user_live.columns = ['source','target']
user_live['type'] = "POSTED"

# Comment -> Video
comment_video = comments[['comment_id','video_id']]
comment_video.columns = ['source','target']
comment_video['type'] = "ON_VIDEO"

# Live -> Channel
live_channel = live[['live_id','channel']]
live_channel.columns = ['source','target']
live_channel['type'] = "ON_CHANNEL"

# Video -> Channel
video_channel = videos[['video_id','channel']]
video_channel.columns = ['source','target']
video_channel['type'] = "BELONGS_TO"

```

```python
# =========================================
# SAVE FILES
# =========================================
users.to_csv(f"{output_folder}/users.csv", index=False)
video_nodes.to_csv(f"{output_folder}/videos.csv", index=False)
comment_nodes.to_csv(f"{output_folder}/comments.csv", index=False)
live_nodes.to_csv(f"{output_folder}/live.csv", index=False)

user_comment.to_csv(f"{output_folder}/user_comment.csv", index=False)
user_live.to_csv(f"{output_folder}/user_live.csv", index=False)
comment_video.to_csv(f"{output_folder}/comment_video.csv", index=False)
live_channel.to_csv(f"{output_folder}/live_channel.csv", index=False)
video_channel.to_csv(f"{output_folder}/video_channel.csv", index=False)

print("✅ Graph data (comments + live) saved!")
```
