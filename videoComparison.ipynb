```python
# =========================================
# 0. CREATE OUTPUT FOLDER
# =========================================
import os

output_folder = "videoComparison"

if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python

# =========================================
# 1. IMPORT LIBRARIES
# =========================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# =========================================
# 2. LOAD VIDEO DATA
# =========================================


videos = pd.read_csv("Data/English_version/English_translated_video_informationt.csv")
df = videos

print("Dataset shape:", df.shape)

```

```python
# =========================================
# 3. BASIC CLEANING
# =========================================
# Convert numeric columns properly
df['view_count'] = pd.to_numeric(df['view_count'], errors='coerce')
df['comment_count'] = pd.to_numeric(df['comment_count'], errors='coerce')
df['video_likes'] = pd.to_numeric(df['video_likes'], errors='coerce')

# Drop missing values
df = df.dropna(subset=['channel','view_count','comment_count','video_likes'])

```

```python
# =========================================
# 4. CHANNEL COMPARISON (NUMERIC)
# =========================================
channel_summary = df.groupby('channel')[['view_count','comment_count','video_likes']].mean()

print("\n===== CHANNEL AVERAGE =====")
print(channel_summary)

```

```python
# =========================================
# 5. TOTAL PERFORMANCE
# =========================================
total_summary = df.groupby('channel')[['view_count','comment_count','video_likes']].sum()

print("\n===== CHANNEL TOTAL =====")
print(total_summary)
```

```python
# =========================================
# 6. TOP VIDEOS (BY VIEWS)
# =========================================
top_videos = df.sort_values(by='view_count', ascending=False).head(10)

print("\n===== TOP 10 VIDEOS =====")
print(top_videos[['channel','Title','view_count']])
```

```python
# =========================================
# 7. BAR GRAPH (AVERAGE COMPARISON)
# =========================================
channel_summary.plot(kind='bar', figsize=(8,5))
plt.title("Average Performance Comparison (Ch-1 vs Ch-2)")
plt.ylabel("Average Value")
plt.xticks(rotation=0)
plt.tight_layout()

plt.savefig(f"{output_folder}/avg_comparison.png")
plt.close()
```

```python
# =========================================
# 8. TOTAL GRAPH
# =========================================
total_summary.plot(kind='bar', figsize=(8,5))
plt.title("Total Performance Comparison")
plt.ylabel("Total Value")
plt.xticks(rotation=0)
plt.tight_layout()

plt.savefig(f"{output_folder}/total_comparison.png")
plt.close()
```

```python
# =========================================
# 9. VIEWS DISTRIBUTION
# =========================================
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='view_count', hue='channel', kde=True)
plt.title("View Distribution by Channel")

plt.savefig(f"{output_folder}/view_distribution.png")
plt.close()

```

```python
# =========================================
# 10. ENGAGEMENT RATE (IMPORTANT METRIC)
# =========================================
# Engagement = (likes + comments) / views
df['engagement_rate'] = (df['video_likes'] + df['comment_count']) / df['view_count']

engagement_summary = df.groupby('channel')['engagement_rate'].mean()

print("\n===== ENGAGEMENT RATE =====")
print(engagement_summary)

```

```python
# =========================================
# 11. ENGAGEMENT GRAPH
# =========================================
engagement_summary.plot(kind='bar')
plt.title("Engagement Rate Comparison")
plt.ylabel("Engagement Rate")
plt.xticks(rotation=0)
plt.tight_layout()

plt.savefig(f"{output_folder}/engagement_rate.png")
plt.close()

```

```python
# =========================================
# 12. SAVE OUTPUT FILES
# =========================================
channel_summary.to_csv(f"{output_folder}/channel_average.csv")
total_summary.to_csv(f"{output_folder}/channel_total.csv")
top_videos.to_csv(f"{output_folder}/top_videos.csv")
engagement_summary.to_csv(f"{output_folder}/engagement_rate.csv")

print("\n✅ Video analysis complete! Files saved in 'videoComparison' folder.")
```
